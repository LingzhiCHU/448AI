```python
state_tracker = []


gamma = 0.9
```

```python
class State:
    ball_x = 0
    ball_y = 0
    velocity_x = 0
    velocity_y = 0
    paddle_y = 0
    terminate = False
    
    
    
    def print_state(self):
        print(np.around(self.ball_x, 2) , " " , np.around(self.ball_y, 2) , " " ,  
              np.around(self.velocity_x, 2) , " " ,  np.around(self.velocity_y, 2) , " " ,  np.around(self.paddle_y, 2))
              
        
    def __init__(self, bx, by, vx, vy, p):
        self.ball_x = bx
        self.ball_y = by
        self.velocity_x = vx
        self.velocity_y = vy
        self.paddle_y = p
        
         

    def up(self):
        self.paddle_y = max(self.paddle_y - 0.04, 0)
        
        #self.update()
        
    def down(self):
        self.paddle_y = min(self.paddle_y + 0.04, 0.8)
        #self.update()
        

    def update(self):
        global state_tracker
        
        self.ball_x += self.velocity_x
        self.ball_y += self.velocity_y
        
        if self.ball_x >= 1:        # hitting right bound
            if self.ball_y >= self.paddle_y and self.ball_y <= self.paddle_y + 0.2:    # hitting paddle
                
                
                
                self.ball_x = 2 - self.ball_x
                
                curr_state = self.match()
                if curr_state not in state_tracker:
                    print(curr_state)
                    state_tracker.append(curr_state)
                
                reward(state_tracker,1)
                
                self.velocity_x = -self.velocity_x + np.random.uniform(-0.015, 0.015)
                if self.velocity_x < 0:
                    self.velocity_x = min(-0.03, self.velocity_x)
                else:
                    self.velocity_x = max(0.03, self.velocity_x)
                if self.velocity_x > 1:
                    self.velocity_x = 1
                elif self.velocity_x < -1:
                    self.velocity_x = -1
                
                self.velocity_y += np.random.uniform(-0.03, 0.03)
                if self.velocity_y > 1:
                    self.velocity_y = 1
                elif self.velocity_y < -1:
                    self.velocity_y = -1
                    
                
                state_tracker = []
                
            else:    # missing paddle
                self.ball_x = 2 - self.ball_x
                #self.match(-1)
                reward(state_tracker,-1)
                state_tracker = []
                self.terminate = True

        else:

                
            
            if self.ball_x <= 0:     # hitting left bound
                self.ball_x = -self.ball_x
                self.velocity_x = -self.velocity_x
            elif self.ball_y <= 0:
                self.ball_y = -self.ball_y
                self.velocity_y = -self.velocity_y
            elif self.ball_y  >= 1:
                self.ball_y = 2 - self.ball_y
                self.velocity_y = -self.velocity_y
                
                
            prev_state = self.match()
            if prev_state not in state_tracker:
                print(prev_state)
                state_tracker.append(prev_state)            
        
    def match(self, value):
        b_x = math.floor(self.ball_x * 12)
        b_y = math.floor(self.ball_y * 12)
        if self.velocity_x > 0:
            v_x = 1
        else:
            v_x = 0    # velocity_x <= 0 map to 0
            
        v_y = 0
        if self.velocity_y > 0 and abs(self.velocity_y) > 0.015:
            v_y = 1
        elif self.velocity_y < 0 and abs(self.velocity_y) > 0.015:    #velocity_y < 0.015 map to 2
            v_y = 2

        p = math.floor(self.paddle_y / (1 - 0.2))
        
        print(b_x, " ", b_y)
        states[b_x * 12 + b_y][v_x][v_y][p] += value
        
    def match(self):
        b_x = math.floor(self.ball_x * 12)
        b_y = math.floor(self.ball_y * 12)
        if self.velocity_x > 0:
            v_x = 1
        else:
            v_x = 0    # velocity_x <= 0 map to 0

        v_y = 0
        if self.velocity_y > 0 and abs(self.velocity_y) > 0.015:
            v_y = 1
        elif self.velocity_y < 0 and abs(self.velocity_y) > 0.015:    #velocity_y < 0.015 map to 2
            v_y = 2

        p = math.floor(self.paddle_y / (1 - 0.2))


        print ( "%d, %d, %d, %d, %d" %(b_x ,b_y, v_x, v_y, p))
        return ((b_x ,b_y, v_x, v_y, p))


        
   
```

```python
    def reward(state_tracker, value):
        sum_util = 0
        for counter, (b_x ,b_y, v_x, v_y, p) in enumerate(state_tracker):
            states[b_x * 12 + b_y][v_x][v_y][p] += (gamma ** counter) * value
            sum_util += states[b_x * 12 + b_y][v_x][v_y][p]
        return sum_util
```
